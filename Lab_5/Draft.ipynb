```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('plots.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
df
```

```python
df['sqrt_vin'] = df['v5']**0.5
df.head()
```

```python
plt.plot(df['v5'], df['vout'], label='simulation V_out v/s V_in')
plt.plot(df['v5'], df['sqrt_vin'], label='theoretical V_out v/s V_in')

plt.title('V_out v/s V_in')
plt.legend()
plt.show()
```

```python
plt.plot(df['v5'], df['vout'], marker='o', linestyle='-', label='Simulation V_out v/s V_in')  # Cross symbol for simulation data
plt.plot(df['v5'], df['sqrt_vin'], marker='x', linestyle='-', label='Theoretical V_out v/s V_in')  # Open circles for theoretical data
plt.title('V_out v/s V_in')
plt.legend()
plt.show()
```

```python
import math
df['theoretical_ln_vout'] = df['sqrt_vin'].apply(lambda x: math.log(x))
df
```

```python
plt.plot(df['ln_vin'], df['ln_vout'], marker='o', linestyle='-', label='Simulation ln(V_out) v/s ln(V_in)')  # Cross symbol for simulation data
plt.plot(df['ln_vin'], df['theoretical_ln_vout'], marker='x', linestyle='-', label='Theoretical ln(V_out) v/s ln(V_in)')  # Open circles for theoretical data
plt.title('ln(V_out) v/s ln(V_in)')
plt.legend()
plt.show()
```
